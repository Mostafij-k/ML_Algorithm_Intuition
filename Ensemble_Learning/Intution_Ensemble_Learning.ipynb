```python
# Ensemble is a collection of multiple machine learning algorithm.
# It's create a big model by combining a group of small model to make better and more accurate prediction.
```

```python
# It is based on wisdom of the crowd.
# Means that a group of people together make better decision than one persion alone.
# Same, many model together make better prediction than one model alone.
```

```python
# Base model of ensemble learning should be different.
# It occur by two ways:
    # Different model same data
    # Same model different data

# Different data and different model may also occur.
```

```python
# Output of Ensemble learning
# For classification use mejority voting count
    # m1=yes, m2=yes, m3=no, m4=yes. Final result= yes (mejority)
# For regression add all ouput value and calculate mean. Final output=mean
    # m1=4, m2=3, m3=4, m4=5. Final result= 4
```

```python
# Type of Ensemble Learning(Important):
    # 1. Voting Ensemble
    # 2. Bagging Ensemble: RandomForest
    # 3. Boosting Ensemble: Adaboosting, Gradient Boosting, Xgboost
    # 4. Stacking Ensemble
```

```python
# Voting Ensemble: 
# Geting same dataset. But base model are different algorithm.
```

```python
# Stackin Ensemble:
# All most same as voting but here give weightage to every model to decide which model is more important
# Here, Also use Final model which use the output of base model.

# Example:
# m1    m2  m3  result
# 1     0   0   0
# 1     1   0   1       # Use the dataset by model 4
# 0     1   1   1
```

```python
# Bagging Ensemble:
    # Bootstrapped Aggregation
    # Same model but different dataset

# Three base model:
    # SVM1   SVM2     SVM3
    # Here, Data will collect randomely from main data
    # Suppose, we have main Data (1000 sample)
    # For base model SVM1 collect D1=500 sample
    # After trainin SVM1, D1=500 sample wiil return to main Data
    # Again collect D2=500 sample for SVM2 and return to main data
    # Again Collect D3=500 and train SVM3
    # Prediction will mejority or mean

# RandomForest is the special case Bagging
```

```python
# Boosting:
    # Same model (Example: LR)
    # Boosting Algorithm minimize error step by step

# Step:
# Data → LR1 (Make error)→ LR2 (Less error Than LR1)→LR3(Less error Than LR2)
```

```python

# Benefit:
    # Improvement of performance
    # Reduce Bias and Variance (Low Bias and Low Variance)
    # DT + Ensemble= Low Bias + Low Variance
    # Robustness

# Disadvantages:
    # Require High Computational Power
```
