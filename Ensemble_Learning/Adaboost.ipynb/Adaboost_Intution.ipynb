```python
            #--------Adaboost------#
# Adaboost is a supervised ensemble learning algorithm used for classification and regression.


# When a machine learning algorithm has accuracy just over 50% called weak learner.
# AdaBoost builds a strong model by combining many weak learners
# Usually a comibination small decision trees
# Each new model pays more attention to previously mis classified data points.
```

```python
# Decision Stumps: A weak learner
# Decision Stumps: A single node (max_depth=1) Decision tree
# Stage Wise Additive method=Sequentially add multiple weak lerner
# Upsampling give weightage to some point and algorithm focus on this point for correct classification.
```

```python
# Adaboost classification will be +1 and -1
# Adaboost is a stage wise additive method
```

```python
# Step
# Train → Error → Alpha → Update weights
```

```python
# Initialize weights
    #Give equal weight to all training samples.
# Train weak model
    # Fit a simple model (usually a decision stump).
# Calculate error
    # Check which samples were misclassified.
# Assign model weight (alpha)
    # Better models get higher weight.
# Update sample weights
    # Increase weights for wrongly classified samples, decrease for correct ones.
# Normalize weights
    # Ensure total weight = 1.
# Repeat steps 2–6
# Build multiple weak learners.
# Final prediction
    # Combine all models using weighted voting.
```

```python
# Equation of Adaboost, 𝐻(𝑥)=sign⁡(∑𝛼𝑡ℎ𝑡(𝑥))
# ht(x) = prediction of weak learner 
#αt = weight of weak learner t (based on its accuracy)
# T = total number of weak learners
# Final output = weighted vote of all learners
```

```python
# Upsampling:
# Means give more weightage to wrong classified point and decrease weitage of correctly classified point.
# Boost of wrong classified value.
```
