```python
# Gradient Boosting:
    # Gradient Boosting is a machine learning technique used for regression and classification.
```

```python
# Features:
    # It builds a strong predictive model by combining many weak models (usually decision trees).
    # Learn step by step each time focusing on what the current model is doing wrong.
    # It is an ensemble
    # Models are trained sequentially
    # Sequential Stage wise addition method
    # For regression loss function will be MSE
    # For Classification loss function will be log loss

# Each time tell how much error occur to forward model and forward model try to solve the error.
```

```python
# Step of Gradient Boosting

# Step:1
    # Prediction of model 1.
    # For regression problem first model will be mean value of output value.
    # Model 1 not a algorithm (DT,svm..)
# step: 2   
    # Check the loss of M1 through loss function
    # Loss funtcion=actual-predicted=pseudoresidual
    # For classification Prediction will e log odds

    # Log(odds)=log(Total number of 1/Total number 0f 0)
    #  Then calculate probabilty, p=1/1-e^-log(odds)
# Step:3
    # Then loss/pseudoresidual transfer to m2
    # Model 2 will be DT (Best for Gradient Boosting)
# Step: 4
    # Input will be same as m1 but output will be residual 1/mistake m1
    # Final prediction= m1+lr* m2
    # Here, add learning rate to avoid overfiting
# Step:4
    # For model 3 residual will be actual-predicted
# Residual2= actual -predicted
        #  = actual-m1+lr*m2

# Through this journey (m1-mn) residual tense to zero
# That is how GB works and predict actual value correctly at the end.
# For GB max leaf node will be 8-32
```
