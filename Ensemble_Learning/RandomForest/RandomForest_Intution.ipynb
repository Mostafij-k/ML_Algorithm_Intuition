```python
            #--------RandomForest---------#
# When the base model of Bagging is a Decision Tree, it is called RandomForest
# It is an ensemble technique
# It is basically a collection of trees
# RandomForest internally a implementation of Bagging technique

# Why Called RandomForest
    # Forest= Collection of Trees (DecisionTrees)
    # Random= Collect sample randomely
    # If Random sampling occurs with Replacement, duplicated sample may occur
    # If Random sampling does not occur with Replacement, duplicated not occur
    # Sampling=Column Sampling, Row Sampling or combination
```

```python
# How RF perform so well?

# Bias Variance Tradeoff and RandomForest
    # The relationship between bias and variance is inversely propotional: Bias ∝ 1/Variance
    # Example of LBHV algorithm: Fully grown DT, SVM, KNN
    # Example of HBLV algorithm: LR
    # RandomForest Convert: LBHV to LBLV

# As RF is a collection of DT, noise pint ditributed into different DT which reduce the impact of noise point.
# Variance becomes low as the impact of noise point reduce.
```

```python
# Code Example of BiasVariance and RandomForest
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Create data and split into train and test set
np.random.seed(42)
X, y = make_circles(n_samples=500, factor=0.1, noise=0.35, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Scatter plot
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
# Show the fully grown DT 

dtree = DecisionTreeClassifier(random_state=42)
dtree.fit(X_train, y_train)

plt.figure(figsize=(12, 8))
x_range = np.linspace(X.min(), X.max(), 100)
xx1, xx2 = np.meshgrid(x_range, x_range)
y_hat = dtree.predict(np.c_[xx1.ravel(), xx2.ravel()])
y_hat = y_hat.reshape(xx1.shape)
plt.contourf(xx1, xx2, y_hat, alpha=0.2)
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis', alpha=.7)
plt.title("Decision tree")
plt.show()
```

```python
# Low Bias and Low Variance using RF
rf = RandomForestClassifier(n_estimators=500, random_state=42)
rf.fit(X_train, y_train)

plt.figure(figsize=(12, 8))
x_range = np.linspace(X.min(), X.max(), 100)
xx1, xx2 = np.meshgrid(x_range, x_range)
y_hat = rf.predict(np.c_[xx1.ravel(), xx2.ravel()])
y_hat = y_hat.reshape(xx1.shape)
plt.contourf(xx1, xx2, y_hat, alpha=0.2)
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis', alpha=.7)
plt.title("Random forest")
plt.show()
     
```

```python
# RandomForest vs Bagging
    # Bagging base model = Any model
    # RandomForest base model= DecisionTree 
```

```python
# Why RF is better tan Bagging?
# Why Bagging with DT not called RF?

# Main difference between bagging and RF is the process of Features sampling
    # In bagging features sampling finalize at tree label
    # On the other hand in RF node label column sampling happen.
    # This process makes RnadomForest better than Bagging.
```

```python
# Code Example

import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
# Make ML Dataset
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5,n_clusters_per_class=1)
     
```

```python
# Create data
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target'] = y
print(df.shape)
df.head()
```

```python
# Create Bagging object
bag = BaggingClassifier(max_features=2)
```

```python
# Fit the model
bag.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
# Create plot to show the internal fetaures
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
# Here, each node contain whether feature[0] or feature[1]
# This two feature selected at first tree label
```

```python
# Create RF object
rf = RandomForestClassifier(max_features=2)
```

```python
# Fit the model
rf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
# Show the tree figures
plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[4])
plt.show()
```

```python
# Here, each nodel contain different feature
# Include all five features
```

```python
# A model is called Random Forest only if:

    # Uses Decision Trees 
    # Uses bootstrap sampling 
    # Uses feature randomness at splits ❗ (this is the defining condition)
    # Missing #3 = NOT Random Forest
```
