```python
# Machine learning method used to predict a continuous value using multiple independent features.
# Contain 1 output(target) variable and more than 1 input variable(independent)
# Equation, y=βo+β1X1+β2X2--------+βnXn, line equation
# try to find the best coefficients (β) that minimize prediction error.
# It minimize ∑i=1n​(yi​−y^​i​)2 (called ordinary least squares)
```

```python

# Import liraries
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
# Create data
X,y = make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise=50)
```

```python
# Cnvert dataframe
df = pd.DataFrame({'feature1':X[:,0],'feature2':X[:,1],'target':y})
```

```python

df.head()
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
# train the model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
# Evaluate the model
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
```
