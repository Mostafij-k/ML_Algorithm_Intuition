```python
                #---------KNN---------#

# KNN finds the K nearest neighbors and makes a decision based on them.

# When a new data point needs to be classified or predicted:
    # The algorithm locates it within the space
    # Measures its distance to all existing data points using a distance metric such as Euclidean distance
    # It then identifies the K closest points, known as the “nearest neighbors.
    # At last perform majority count 
    # Pick the class with the highest count (majority)
    # Assign that class to the new point

# As KNN distance based algorithm perform scaling before training
```

```python
# How to select K value?

    # Rule of Thumb:
        # A common starting point: K≈ root over (N)
        # where N = number of training samples
        # If N=100, then K≈10

    # Use Cross-Validation (Best Method)
        # Try multiple K values and pick the one with highest accuracy.

    # Avoid even K (for classification)

# Very small K	Overfitting (too noisy)
# Medium K	Balanced (best choice)
# Very large K	Underfitting (too smooth)
```

```python
# Decision Boundary (Decision Surface)
    # It is use to understand the performance of classification Algorithm (KNN, SVM, LR, DT). 
    # For binary classification it devide the space into two region.
    # Decision surface = Create regions where KNN predicts different classes
```

```python
# Overfitting and Underfitting in KNN

# Overfitting
    # What happens?
    # The model learns the training data too perfectly, including noise and outliers.

    # When does it occur?
    # Very small K
    # The model relies on very few neighbors
    # Even noisy points strongly influence predictions

# Undefitting:
    # The model is too simple and fails to capture patterns in the data.
    # Very large K
    # Too many neighbors are considered
```

```python
# Limitations: 
# KNN is called a lazy learning algorithm because it does not learn a model during training
# It simply stores the data and delays all computation until prediction time
# It does not work without any query. 
    # Not suitable for large dataset
    # Not suitable for high dimension data
    # In high dim distance not reliable but KNN totaly distance base
    # KNN not work well with outliers
    # Not work with non-homogenous scaling data
    # Imbalanced dataset
```
