```python
                #---------Hold-Out Method--------#
# The hold-out method is one of the simplest techniques for evaluating the performance of a machine learning model.

# It involves splitting your dataset into separate training and testing sets.
# The model is trained on the training set and then evaluated on the testing set.

# Step of the Hold-Out Method
# Split the dataset:
    # Divide the dataset into two (or sometimes three) subsets:

    # Training set: Used to train the model (usually 60–80% of data).
    # Testing set: Used to evaluate the model (usually 20–40% of data).
    # Sometimes, a validation set is also used if hyperparameter tuning is required.
# Train the model:
    # Fit the machine learning model using only the training data.
# Evaluate the model:
    # Test the model’s performance on the unseen testing data using appropriate metrics
```

```python
# Code Example:
# Import required Libraries
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_iris
Data=load_iris()
Data.data
Data.target
```

```python
# Create input and output dataset
X=Data.data
y=Data.target
# Split the dataset
X_train,X_test, y_train, y_test=train_test_split(
    X,y,test_size=0.3
)
```

```python
# Perform logisticregression
# Create LogisticRegression object
lr=LogisticRegression(solver='lbfgs', max_iter=1000)
# Fit the model
lr.fit(X_train, y_train)
# Predict the model
y_pred=lr.predict(X_test)
# Find the accuracy score
accuracy_score(y_test,y_pred)
```

```python
# Perform SVM
# Create SVM object
svc=SVC(gamma='auto')
# Fit the model
svc.fit(X_train, y_train)
# Predict the model
y_pred1=svc.predict(X_test)
# Find the accuracy score
accuracy_score(y_test,y_pred1)
```

```python
# Perform RandomforestClassifier
# Create RandomforestClassifier object
rf=RandomForestClassifier(n_estimators=40)
# Fit the model
rf.fit(X_train, y_train)
# Predict the model
y_pred2=rf.predict(X_test)
# Find the accuracy score
accuracy_score(y_test,y_pred2)
```
