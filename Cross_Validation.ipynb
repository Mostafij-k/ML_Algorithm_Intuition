```python
                # --------Cross Validation-------#
# Cross Validation (CV) is a technique to evaluate how well your machine learning model will perform on unseen data.

# Instead of splitting the dataset just once into
# Split it multiple times and test the model multiple times.

# Most Common Type: K-Fold Cross Validation
# How it works:
    # Divide dataset into K equal parts (folds)
    # Train on K-1 folds
    # Test on remaining 1 fold
    # Repeat until every fold has been used as test set
    # Take the average score

# Reduces overfitting risk
# Uses entire dataset for both training & testing
# Gives stable performance estimate
# Very useful when dataset is small
```

```python
# Code Example
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
# Import Dataset
Df=pd.read_csv('titanic.csv', usecols=['Age','Pclass','SibSp','Parch','Survived'])
```

```python
# BAsic info
Df.info()
```

```python
# Drop null value from age column
Df.dropna(inplace=True)
Df.info()
```

```python
# Create input and output set
X=Df.drop(columns=['Survived'])
y=Df['Survived']
```

```python
# Train the model

# Create a model object
model=LogisticRegression()
```

```python
# Perform logistic regression using cross validation
scores = cross_val_score(model, X, y, cv=5)
print(scores)
```

```python
# Print average score
print("Average accuracy:", scores.mean())
```
