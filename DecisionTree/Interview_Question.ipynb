```python
# Entropy vs Gini impurity?
    # Both are metrics used to measure impurity
    # Entropy measures the level of disorder using logarithms
    # Gini impurity measures the probability of misclassification.
# Entropy use in ID3 and Gini impurity use in CART algorithm
```

```python
# What is ID3?
# Iterative Dichotomiser 3 is a decision tree algorithm used for classification tasks.
# Selecting the feature that gives the highest Information Gain

# How It Works (Step-by-Step)
    # Start with the full dataset
    # Calculate Entropy of the dataset
    # For each feature:
    # Compute Information Gain: E(parent)-(weighted average)* E(children)
    # Choose the feature with highest Information Gain
    # Split the dataset based on that feature
    # Repeat recursively for each subset
    # Stop when:
    # All data is pure, OR
    # No features remain
```

```python
# What is CART?
# CART (Classification and Regression Trees) is a popular decision tree algorithm
# Used for both:
    # Classification (Predicting Categories)
    # Regression (Predicting Continious Values)

# Splitting the dataset into two parts (binary splits) at each step
# Choosing splits that make the data as pure as possible

# Splitting Criteria
# For classification → uses Gini Impurity
# For regression → uses variance reduction (MSE)

# How it work
    # Start with the full dataset
    # Try all possible splits on all features
    # Choose the split with lowest impurity (Gini or MSE)
    # Split into two nodes
    # Repeat recursively
    # Stop when:
    # Maximum depth reached
    # Minimum samples per node reached
```

```python
# When should use entropy and when should use Gini

# Gini Impurity:
    # Faster Computation
    # Large Dataset
# Entropy:
    # Smaller Dataset
```

```python
# How decesion tree regression work?
    # A Decision Tree Regressor is used when the target variable is continuous.

# Core Idea:
# Split the data into regions such that:
    # Values within each region are as similar as possible
# Final prediction = average (mean) of values in that region

# How It Works (Step-by-Step)
# Start with full dataset
# Try all possible splits:
    # For each feature:
    # Try different split points (X<5, X<10)
    #  
# Choose best split: Select split that minimize error (Lower MSE)
# Split into two regions
# Repeat recursively until stopping criteria are met.
# Assign prediction: mean of target values in that node
```

```python
# Decision Tree Working For Categorical and Numerical Features?
# The tree:
    # Evaluates ALL possible splits
    # (both numerical thresholds + categorical groupings)
    # Then selects the one with:
    # Lowest Gini impurity (classification)
    # Lowest MSE (regression)

# Example:
# Suppose you are predicting cancer risk:
# Possible splits:
# Age < 50
# Gene expression > 2.1
# Gender = Male vs Female
# Tumor type ∈ {A, B} vs {C}
    # The algorithm compares all of these and picks the best one.
```
