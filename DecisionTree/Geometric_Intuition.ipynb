```python
# Decision tree is nothing but a giant structure of nested if else condition
# Mathematically speaking, decision trees use hyperplanes which run parallel to any one of the axes to cut your coordinate system into hypercuboids.
# Decision tree can work in both numerical and categorical column
```

```python
# Advantages:
# Intuitive and Easy to understand
# Minimal data preparation is required

# Disadvantages:
# Overfitting
# Prone to error for imbalanced dataset
```

```python
# Pseudocode:
    # Begin with your training dataset, which should have some feature variables and classification or regression output.
    # Determine the 'best feature' in the dataset to split the data on more on how we define 'best feature' later.
    # Split the data into subsets that contain the correct vlaues for this best feature. 
    # This spliting basically defines a node on the tree each node is a spliting point based on a certain feature from our data.
# First node called root node
# Those point data will spilt called spliting point
# last node called leaf node
# Between root node and leaf node called decision node

# Note: Decision tree algorithm mostly use in classification  problem. 
# The logic  of decision trees can also be applied to regression problem, hence called CART
```

```python
# How it work?

# Entropy:
    # Entropy is nothing but measure of disorder. or measure of purity or impurity.
    # More knowledge less entropy
    # Is a measure of impurity (or randomness) in a dataset.
# Formula:
# H(S)=−∑i=1 to c [​pi​log2​(pi​)]
# H(S): Entropy of dataset 𝑆 (measure of impurity)
# 𝑝𝑖: Probability of class 𝑖
# c: Number of classes

# Entropy is zero at leaf node
```

```python
# Role in Decision Trees
# A decision tree uses entropy to decide where to split.
# How?
    # Calculate entropy of current node
    # Try different splits
    # Choose the split that:
    # reduces entropy the most

    # This reduction is called Information Gain
```

```python
# Information Gain:
    # Information Gain, is a metric used to train Decision Trees. 
    # Specifically, this metric measures the quality of a spilt.
    # The information gain is based on the decrease in entropy after a data-set is split on an feature/column.
    # Constructing a decision tree is all about finding attribute/feature that returns highest information gain.

# Information Gain=E(parent)-(weighted average)* E(children)
```

```python
# Step to find decision tree

# Step-1:
    # Findout the entropy of parent node.
    # Example: 0.94
# Step-2:
    # Calculate the entropy of each children node
    # Example:0.97
# Step-3:
    # Calculate the weighted entropy of children
    # Example: .69
# Step-4:
    # Calculate the information gain
    # IG=0.94-0.69 =0.28
# Step-5:
    # Calculate information gain for all the columns
# Whichever column has the higest information gain (maximum decrease in entropy) the algorithm will select that column to split the data.
# Find the information gain recursively
```

```python
# Gini Impurity:
# Gini Impurity is a metric used in decision trees to measure how “impure” or “mixed” a node is.
# Intuition: 
    # If a node contains only one class → it is pure → Gini = 0
    # If a node contains a mix of classes → it is impure → Gini > 0
    # The more mixed the classes → the higher the Gini impurity
#  Gini=1−∑i=1 to n (​pi2)​
# Pi= probability of class i
# n = number of classes
# Decision tree try to Minimize Gini impurity after splitting

# Gini vs Entropy:
    # Both measure impurity
    # Gini is faster to compute (no log)
    # Entropy is more sensitive but slower
```
