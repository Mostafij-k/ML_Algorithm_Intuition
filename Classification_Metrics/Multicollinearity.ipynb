```python
# Multicollinearity is a phenomena where two or more independent variables are highly correlated.
# This creates redundant information, skewing the results in a regression model.
# Correlation may be positive or negative
# Multicollinearity happens when predictor variables contain similar information.

# Why bad?
# when multicollinearity exists, the coefficients become very sensitive, and their values can change drastically.
# Do not understand the contribution of different feature

# Is always bad?
# No, If your goal is prediction accuracy, multicollinearity may not be harmful.
```
