```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Import dataset
df=pd.read_csv('Mall_Customers.csv')
df.shape
```

```python
# Show the dataset
df.head(3)
```

```python
# Draw a scatter plot (use last two column)
x=df.iloc[:, 3].values
y=df.iloc[:, 4].values
plt.figure(figsize=(4,4))
plt.scatter(x,y)
```

```python
#  Extract input column
X=df.iloc[:, 3:]
X.shape
```

```python
# Trin the model with different k value and calculate WCSS

wcss=[]
for i in range(1,11):
    km=KMeans(i)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
# Now create a plot to sohow elbow point
# Best k value findout
plt.figure(figsize=(5,5))
plt.plot(range(1,11), wcss)
```

```python
# Best k value will be 5 (elbow point)
```

```python
# Now again train the model with best k=5

# Create a kmeans object
km=KMeans(n_clusters=5)
y_cluster=km.fit_predict(X)
```

```python
# Show the y_cluster
X[y_cluster==4].shape
print(set(y_cluster))
```

```python
# Now create a scater plot to visualize cluster
colors = ['red', 'green', 'blue', 'yellow', 'black']
for i in range(5):
    plt.scatter(X.iloc[y_cluster==i, 0], X.iloc[y_cluster==i, 1], color=colors[i])
```
