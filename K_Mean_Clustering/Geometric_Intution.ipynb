```python
            #---------K-mean Clustering------#
# K-mean clustering is a unsupervised machine learning algorithm.
# For higher dimentional dataset it will be the best choice
```

```python
# Step of K mean clustering algorithm

# 1. Choose the number (n) of clusters (K)
# 2. Initialize centroids (centroid= n cluster)
    # Algorithm randomely select any point as centroid
# 3.Assign data points to clusters
    # Calculate the distance (usually Euclidean) to all centroids
    # Assign the point to the nearest centroid
# 4. Update centroids
    # New centroid will be the center of a cluster
    # Calculate the mean value of x and y axis of all point in a cluster 
# 5. Repeat Step 3 and Step 4 
# 5. Check for Finish (When centriod not move from before position finish the work)

# Note: K-means tries to minimize the Within-Cluster Sum of Squares (WCSS)
```

```python
# How can we choose the best number of cluster?
 
    # To find the best value of K in K-mean clustering use Elbow method.

# Behind Elbow method:
    # When increase K the clusters become tighter.
    # The Within-Cluster Sum of Squares (WCSS) decreases.
    # After a certain point(elbow point), the improvement becomes very small.
    # In graph  x=number of cluster, y=WCSS
```

```python
# What is WCSS?
    # It is used to evaluate how 'tight'/ 'compact' the clustes. 
    # WCSS calculates the total squared distance between each data point and centroid.
    # If points are very close to their centroid → WCSS is small.
    # If points are spread out → WCSS is large.
    # Lower WCSS = better clustering.

# Increase cluster=decrease WCSS
# But after certain point it will not decrease (elbow point)
# So you need to finout elbow point
# WCSS will be zero when each point work as centroid

# So when cluster is one WCSS will be high
# WCSS1>WCSS2>WCSS3
```

```python
# 1. Do you need to scale data in k_mean clustering?
    # Yes. it’s usually essential for K-means clustering.
```

```python
# 2. How do you handle outlier in k-mean clustering?
    # Outliers have a bad impact on k_mean clustering
    # Centroid is very sensative to outlier
    # Outliers can pull centroids toward them which create poor cluster
    # Remove them or use robust scaling method to handle outlier
```

```python
# 3. Why k-mean algorithm consider as greedy algorithm? short anser
    # K-means is considered greedy because it makes locally optimal decisions at each step without considering the global optimum.
```

```python
# 4. Explain significance of elbow method?
    # Elbow method helps to choose the optimal number of clusters (k) in K-means.
    # It plots within-cluster sum of squares (WCSS) vs. number of clusters to find elbow point
```

```python
# 5. What is the curse of dimensionality and how does it effect in k-mean clustering?
    # Meaning the number of features (dimensions) increases, the data becomes sparse and distance measures become less meaningful.
    # K-means depends on distance 
    # High dimention may give poor cluster
```

```python
# 6. Can you use k-means clustering for high-dimensional data?
    # K-means can be used, but for high-dimensional data, should reduce dimensions first to get reliable results.
    # Apply dimensionality reduction (e.g., PCA) before K-means
    # Use feature selection to keep only important variables
```

```python
# What is silhouetee and how it is used in k-mean clustering    
    # Silhouette (score) measures how well each data point fits within its cluster in K-means.
    # Compare average distance to its own cluster (a)
    # Compare average distance to nearest other cluster (b)
    # Close to +1 → well clustered
    # Around 0 → overlapping clusters
    # Close to -1 → wrongly assigned
```

```python
# How can the k-mean clustering be optimized for large dataset? 
    # Use Mini-batch K-means + good initialization + reduced dimensions for best performance on large data.
```
