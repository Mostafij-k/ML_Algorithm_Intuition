```python
            #---------K-mean Clustering------#
# K-mean clustering is a unsupervised machine learning algorithm.
# For higher dimentional dataset it will be the best choice
```

```python
# Step of K mean clustering algorithm

# 1. Choose the number (n) of clusters (K)
# 2. Initialize centroids (centroid= n cluster)
    # Algorithm randomely select any point as centroid
# 3.Assign data points to clusters
    # Calculate the distance (usually Euclidean) to all centroids
    # Assign the point to the nearest centroid
# 4. Update centroids
    # New centroid will be the center of a cluster
    # Calculate the mean value of x and y axis of all point in a cluster 
# 5. Repeat Step 3 and Step 4 
# 5. Check for Finish (When centriod not move from before position finish the work)

# Note: K-means tries to minimize the Within-Cluster Sum of Squares (WCSS)
```

```python
# How can we choose the best number of cluster?
 
    # To find the best value of K in K-mean clustering use Elbow method.

# Behind Elbow method:
    # When increase K the clusters become tighter.
    # The Within-Cluster Sum of Squares (WCSS) decreases.
    # After a certain point(elbow point), the improvement becomes very small.
    # In graph  x=number of cluster, y=WCSS
```

```python
# What is WCSS?
    # It is used to evaluate how 'tight'/ 'compact' the clustes. 
    # WCSS calculates the total squared distance between each data point and centroid.
    # If points are very close to their centroid → WCSS is small.
    # If points are spread out → WCSS is large.
    # Lower WCSS = better clustering.

# Increase cluster=decrease WCSS
# But after certain point it will not decrease (elbow point)
# So you need to finout elbow point
# WCSS will be zero when each point work as centroid

# So when cluster is one WCSS will be high
# WCSS1>WCSS2>WCSS3
```
