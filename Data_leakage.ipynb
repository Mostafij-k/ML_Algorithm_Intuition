```python
# Data leakage = the model learns something it should not know during training.
# Data leakage in machine learning occurs when a model uses information during training that wouldn't be available at the time of prediction.
# The model will perform well in testing and validation, but when used in production, it becomes entirely inaccurate.
```

```python
# There are two types of leakage: target leakage and train-test contamination.
# Target leakage: Models include data that will not be available when the model is used to make predictions.
# Train-test contamination: When both training and validation data are used to create a model, often due to improper splitting or preprocessing.
```

```python
# Data leakage can be occurs due to a variety of factors such as:
# Inclusion of future information
# Inappropriate feature selection
# External data contamination
# Data preprocessing errors
# Incorrect cross-validation
# Normalization
# Validation and process changes leakage
```
