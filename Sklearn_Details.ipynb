```python
# Estimator: An Object that can learn from data and has a fit method
# Example: 
    # 1. Linear_fit
    # 2. KNN_fit
    # 3. Scalar_fit
# Note: fit learn from data
    # For linear regression learn weight
    # For standardscalar scaling learn mean/median
```

```python
# Estimator Type ;
    # There are two type of estimator 1. Predictor 2. Transformer
# Predictor: estimator that make prediction when new data comes
    # contain .predict() function
    # lr, SVM, DT, XgBoost etc. algorithm
# Transformer: It is an estimator that make transformation when new data comes
    # Contain .transform()/.fit_transform()
    # ohe, OE, SS, MMS, PCA

# Note: Clustering Algorithm contain only fit method. It has neither .predict() not .transform()
```

```python
# Predict Estimator: 2 type
    # Predict() regression: give numerical outpu
    # predict() classification: predict( 1,0), predict_prob(0.3, 0.2)
```

```python
# Transformer: 
# In scikit-learn, a transformer is a specific type of estimator that is used for transforming datasets.

# Transformers are designed to pre-process data, which can be include tasks such as:
# Scaling, Categorical Data Handling, Handling missing value, and features extraction.
# The main goal of a transformer is to modify or create new features from the original datsets in a way..
# that makes the data more suitable for modeling by machine learning algorithms.

# Contains a fit and transform (fit_transform) method
# Example: Standardization, MinMixScaler, OHE, OE, SimpleImputer, PCA

# Transform Estimator-> Given Data-> Transform-> More Suitable Data for ML Algorithm

# Disadvantage: 
    # Can be  applied only to the entire dataset
```

```python
# Composite Transformer: 
# A composite transformer refers to a transformer that is built from multiple transformers
# or estimators, combining their functionalities to apply a series of transformations or processing steps in a specific way.

# Types:
    # 1. Column Transformer
    # 2. Pipeline
    # 3. Feature Union
```
