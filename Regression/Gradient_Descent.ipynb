```python
# Gradient Descent is an optimization algorithm 
# used in machine learning to minimize the error (loss function)
# Formula, θnew​=θold​−η*∂J(θ)​/∂θ
# Mn=Mo-η* slope
# Bn=Bo-η* slope
# θ = parameter (weight/bias)
# J(θ) = loss function
# ∂J/∂θ= gradient (slope)
# η = learning rate

# Learning rate determines how big a step Gradient Descent takes during parameter updates.
```

```python
# Steps of Gradient Descent
# 1. Initialize parameters randomly
# Example:
# w=0.5
# 2. Make predictions
# 3. Calculate loss
# Example:
# Mean Squared Error (MSE)
# 4. Compute gradients
# 5. Update parameters
# 6. Repeat until minimum loss achieved
```

```python
# Type of GD
    # Batch GD
        # Uses the entire dataset in each update.
        # Stable convergence
        # Slow for large dataset
    # Stochastic GD
        # Uses one training sample at a time.
        # Works well for huge datasets
    # Mini Batch GD
        # Uses small batches of data.
        # Most commonly used in deep learning.
```
