```python
# What is Support Vector Machine?
    # A Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression problem.
    # It is most commonly used for classification problems. Particularly best for binary classification
```

```python
# 1. Main Difference between hard margin and soft margin SVM?

    # Hard Margin SVM assumes that the data is perfectly separable and does not allow for any misclassifications.
    # It works best with clean, noise-free data otherwise it would either fail or overfit.

    # Soft Margin SVM allows some misclassifications by introducing a penalty for points that violate the margin
    # It balances maximizing the margin with minimizing classification errors, controlled by the regularization parameter C.
```

```python
# 2. Stack Variable in SVM?

    # Slack variables in SVM allow the model to accept some mistakes when the data can't be perfectly separated. 
    # With slack variables, SVM can still find the best boundary by balancing between having a wide margin and minimizing errors
```

```python
# For N dimensional data set what is the minimum possible number of Support Vectors?
    # N+1
```

```python
# Name some advantages of SVM
    # Works well in high-dimensional data
    # Effective for small datasets
    # Uses only important data points (Support Vectors)
    # Can handle non-linear data using kernels
    # Good performance in classification problems
    # Less overfitting
```

```python
# What are Support Vectors in SVMs?
    # Support vectors are the data points that lie closest to the decision boundary

# Why Important?
    # Define the position of the hyperplane
    # Control the margin
    # Directly influence the model
```

```python
# What happens when there is no clear Hyperplane in SVM?
    # it adapts in two smart ways:
    # 1. Soft Margin SVM (allow some errors)
        # Allows some misclassification
        # Tries to keep the margin as large as possible while tolerating errors

        # Controlled by parameter C:

        # Large C → fewer errors, tighter margin (risk of overfitting)
        # Small C → more errors allowed, wider margin (better generalization)

    # 2. Kernel Trick (transform to higher dimension)
```

```python
# Why would you use the Kernel Trick?
    # Use to separate non linear dataset by transforming data into higher dimentional spaces
```

```python
# Compare K-Nearest Neighbors (KNN) and SVM?
     # KNN classifies based on nearby data points 
     # SVM finds the optimal boundary that separates classes with maximum margin.
```

```python
# Compare SVM and Logistic Regression in handling outliers
    # SVM:
    # Less sensitive to outliers
    # Only cares about support vectors (points near the boundary)
    
    # Logistic Regression:
    # More sensitive to outliers
    # Uses all data points to fit the model
```

```python
# Provide an intuitive explanation of Linear Support Vector Machines (SVMs)
    # It tries to draw a straight line (or flat plane) that separates two classes as clearly as possible
    # keeps the largest possible distance from both groups
```

```python
# What are Polynomial Kernels?
    # Polynomial kernels enable SVM to model non-linear relationships by mapping data into a higher-dimensional space using polynomial functions.
```

```python
# What is the Hinge Loss in SVM?
    # Hinge loss is the loss function used in Support Vector Machines (SVMs) to measure how well the model separates classes.
    # Hinge loss penalizes points that are misclassified
    # Encouraging SVM to maximize the margin.
```

```python
# When SVM is not a good approach?
    # SVM is not suitable for very large, noisy, highly overlapping, or real-time datasets, and it requires careful tuning to perform well.
```

```python
# When would you use SVMs over Random Forest and vice-versa?
    # Use SVM for small-to-medium, high-dimensional, well-structured data
    # Random Forest for large, noisy, real-world tabular data with mixed features.
```

```python
# How does the value of Gamma affect the SVM?
    # Gamma controls how strongly each data point influences the decision boundary
    # high gamma leads to complex overfitting, while low gamma produces a smoother, more general model.
```

```python
# What are Radial Basis Function Kernels?    
    # RBF kernel transforms data using a distance-based similarity function that allows SVM to create highly flexible non-linear decision boundaries.
```

```python
# What are C and Gamma (γ) with regards to a Support Vector Machine?

    # C (Regularization);
    # C controls how much the model cares about misclassification.

    # Gamma (Y):
    # Gamma controls how far a single training point influences the decision boundary.
```

```python
# When would you use SVM vs Logistic regression?
    # Use Logistic Regression when you need speed, simplicity, and interpretability, 
    # Use SVM when you need higher accuracy on complex or high-dimensional data.
```

```python
# How to select Kernel for SVM?
    # Choose a kernel based on data complexity: start with linear, use RBF for most non-linear problems, and use polynomial only when feature interactions are important.
```
