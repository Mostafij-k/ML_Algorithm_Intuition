```python
            #--------------SVM-----------#
# It’s all about finding the best boundary that separates data points.

# Imagine you have two classes of data points in 2D space.
#  SVM tries to draw a line that separates them.

# In 2D → line
# In 3D → plane
# In higher dimensions → hyperplane

# There are many possible separating lines, but SVM chooses the best one.
# SVM selects the hyperplane that maximizes the margin.
# Margin = distance between the hyperplane and the closest points.
# Margin=Distance beteen positive and negative hyperplane
# Closest point called support vector
# The distnace between positive and negative hyperplane from center hyperplane are equal.

# The decision boundary depends only on support vectors, NOT all data points.

# The goal of SVM is to findout the Hyperplane (Decision boundary) that will maximize distance between positive and negative hyperplane.


# Advantages:
    # Robust to outlier
    # Can separate non linear dataset using kernel
    # Useful in both classification and regression
```

```python
# Geometric Steps of SVM

# Stpe1: Main hyperplane (decision boundary)
    # SVM finds a central separating hyperplane: wTx+b=0

# Step2: Two parallel hyperplanes (margins)
    # Then SVM defines two parallel hyperplanes on either side: wTx+b=1, wTx+b=-1

# It simultaneously finds all three by solving an optimization problem

# The goal is to position them so that:
    # The middle one separates classes
    # The outer ones touch the closest data points (support vectors)
    # The distance between outer planes is maximum
```

```python
# Mathmatical formula
wTx+b=0
# w → weight vector (direction of the hyperplane)
# b → bias (shift)

# SVM tries to maximize:
# Distance between two parallel planes:
wTx+b=+1
wTx+b=−1
# Margin=2​/∥w∥

# Smaller ∣∣w∣∣ → larger margin
# Optimization goal → minimize ∣∣w∣∣
```

```python
# The Uquation of Hyperplane is: wTx=0, when line go through the origin

# wTx= w.x=|W|.|X| cosθ=0
# θ=90, so w is the perpendicular to the plane, Hyperplane
```

```python
# Mathmatics of Support Vector (Hard margin SVM)

    # Hard Margin SVM is possible when data is perfectly separable by linear line
    # Optimization function for hard margin SVM, Maximize, argmax, 2​/∥w∥, such that yi(wTxi+b)≥1
    # Here, yi(wTxi+b)≥1 is constrain for optimization
    # At margin all data point classified perfectly. No misclassification occur

# Decision boundary rule
    # w.u+b≥0, Here, u is the unknow vector (point) for classification.
    # If greater than 0 it will be fall under positive area.
    # If less than 0 it will be fall under negative area.

# Suppose decision boundary equation, 2x+3y+3=0, if compare with w1x1+w2x2+b=0
# w=(2,3), if new vector u=(3,2). So position will be=  2.3+3.2 =positive

# for any value xi
# Predicted y= { +1 if w.xi+b ≥0
#               -1 if w.xi+b <0 }


# So, the goal of the SVM is to find such w and b which will maximize the margin.
# Center (Exactly middle) Hyperplane equation, wTx+b=0
# Positive Hyperplane equation, wTx+b=1
# Negative hyperplane equation, wTx+b=-1


# Constrain of Hyperplane equation, 
# For positive equation, w.xi+b ≥1
# For negative equation, w.xi+b ≤-1

```

```python
# RBF kernel:
# It is a function that transforms data into a higher-dimensional space so that non-linear data can be separated linearly.
# It is also known as the Gaussian kernel.
# RBF kernel = measure of similarity

# Formula
    # 𝐾(𝑥,𝑥′)=exp(−𝛾∥𝑥−𝑥′∥2)
    # x,x′ = two data points
    # ∥𝑥−𝑥′∥2= squared Euclidean distance
    # γ = parameter that controls how far the influence of a single point reaches

 # Role in SVM
    # maps data into higher dimension
    # lows SVM to create non-linear decision boundaries
```

```python
# Hyperparameter in RBF
    # Gamma and C

# Low γ (small value)
    #Far points are considered similar
    #Smooth decision boundary
# High γ (large value)
    # Only very close points influence each other
    # Complex boundary

# C 
# Controls trade-off between margin and classification error

# Small C
    # Allows some misclassification
    # Wider margin

# Larger C
    # Tries to classify all points correctly
    # Narrow margin
```

```python
# Plynomial Kernel
# It transforms data into a higher-dimensional space where the relationship between features can be expressed as a polynomial equation.
# This allows SVM to draw curved decision boundaries.

# Formula
# 𝐾(x,x′)=(γx⋅x′+r)d
# x⋅x′= dot product of two vectors
# γ = scaling factor
# r = constant term (also called coef0)
# d = degree of the polynomial

# Hyperparameter
# D:
# Controls the complexity of the model
# High : Very complex boundary (risk of overfitting)
# gamma:
# Controls influence of features (like RBF)
# High: Complex Boundary
# C:Regularization
# Small C → simpler model
#Large C → strict fitting
# coef:
# Controls the importance of higher-order vs lower-order terms
# model = SVC(kernel='poly', degree=3, gamma=0.1, coef0=1, C=1)
```
